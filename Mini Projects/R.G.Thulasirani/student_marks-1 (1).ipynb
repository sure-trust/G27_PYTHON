# Data Export
 Here we save the existing DataFrame to a CSV file named "student_marks.csv" while including the index.
 This code uses the "to_csv" method of the DataFrame to export the data to a CSV file. We set "index=True" to include the index in    the CSV file.

```python
import pandas as pd
import numpy as np

num_students = 80
num_subjects = np.random.randint(6, 9)  

data = {
    'Class': [],
    'Student_ID': [],
    'Name': []
}
for i in range(1, num_students + 1):
    data['Class'].append('Class A')
    data['Student_ID'].append(i)
    data['Name'].append(f'Student_{i}')
    for j in range(1, num_subjects + 1):
        data[f'Subject_{j}'] = np.random.randint(0, 101, num_students)


df = pd.DataFrame(data)

for _ in range(30):  
    row_idx = np.random.randint(0, num_students)
    col_idx = np.random.randint(3, df.shape[1])  
    df.iat[row_idx, col_idx] = np.nan

df
```

# Machine Learning Model
Here we build a machine learning model to predict a student's performance in "Subject1" based on the marks in other subjects (Subject2, Subject3, etc.). We use a simple linear regression model. 
We split the data into training and testing sets using train_test_split.
We create a Linear Regression model, fit it to the training data, and make predictions on the test data.
Finally, we evaluate the model's performance using Mean Squared Error (MSE)

```python
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x=data['Subject1']
y=data['Subject2']
```

```python
x = np.array(x).reshape(-1,1)
y= np.array(y).reshape(-1,1)
plt.scatter(x,y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
y_pred = reg.predict(x_test)
y_pred
```

```python
y_test
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x))
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x)
```

```python
reg.fit(x_poly,y)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(poly.fit_transform(x)))
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```
