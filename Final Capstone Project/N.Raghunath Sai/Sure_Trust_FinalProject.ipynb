```python
p1.drop("default",axis=1,inplace=True)
p1
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
l1=p1["job"]
l11=le.fit_transform(l1)
l11.reshape(-1,1)
p1["job"]=l11

l2=p1["marital"]
l22=le.fit_transform(l2)
l22.reshape(-1,1)
p1["marital"]=l22

l3=p1["education"]
l33=le.fit_transform(l3)
l33.reshape(-1,1)
p1["education"]=l33

l4=p1["housing"]
l44=le.fit_transform(l4)
l44.reshape(-1,1)
p1["housing"]=l44

l5=p1["loan"]
l55=le.fit_transform(l5)
l55.reshape(-1,1)
p1["loan"]=l55

l6=p1["contact"]
l66=le.fit_transform(l6)
l66.reshape(-1,1)
p1["contact"]=l66

l7=p1["month"]
l77=le.fit_transform(l7)
l77.reshape(-1,1)
p1["month"]=l77

l8=p1["poutcome"]
l88=le.fit_transform(l8)
l88.reshape(-1,1)
p1["poutcome"]=l88

l9=p1["y"]
l99=le.fit_transform(l9)
l99.reshape(-1,1)
p1["y"]=l99

p1
```

```python
p1.info()
```

```python
Y=p1["y"]
Y
```

```python
X=p1.drop("y",axis=1)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import r2_score
LR=LogisticRegression()
LR.fit(X_train,Y_train)
y_pred=LR.predict(X_test)
print(y_pred)
print(Y_test)
print('Mean Absolute Error: ', metrics.mean_absolute_error(Y_test,y_pred))
print('Hean Squared Error: ', metrics.mean_squared_error(Y_test,y_pred))
r2 = r2_score(Y_test, y_pred)
print(f"R2 Score: {r2}")
```
